```python
import os
from pathlib import Path
import matplotlib.pyplot as plt

data_path = Path("data")
class_counts = {}

for style_folder in sorted(data_path.iterdir()):
    if style_folder.is_dir():
        count = len(list(style_folder.glob("*")))
        class_counts[style_folder.name] = count

# Plot it
plt.figure(figsize=(12, 6))
plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=45, ha='right')
plt.xlabel("Architectural Style")
plt.ylabel("Number of Images")
plt.title("Images per Style")
plt.tight_layout()
plt.show()

print(f"Total images: {sum(class_counts.values())}")
print(f"Smallest class: {min(class_counts, key=class_counts.get)} ({min(class_counts.values())})")
print(f"Largest class: {max(class_counts, key=class_counts.get)} ({max(class_counts.values())})")
```
