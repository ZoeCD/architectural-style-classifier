```python
import requests
import os
```

```python
image_path = "../example_images/art-deco.JPG"

if not os.path.isfile(image_path):
    print(f"File does not exist: {image_path}")
    exit(1)
else:
    print(f"File exists: {os.path.abspath(image_path)}")
```

```python
# Health response
response = requests.get(
    "https://architectural-style-classifier.onrender.com/health", timeout=5
)
print(response.json())
```

```python
with open(image_path, "rb") as img_file:
    response = requests.post(
        "https://architectural-style-classifier.onrender.com/predict",
        files={"file": (os.path.basename(image_path), img_file, "image/jpg")}
    )

try:
    print("JSON:", response.json())
except Exception:
    print("Status code:", response.status_code)
    print("Raw response:", response.text)
```
